```python
import pandas as pd
import os
df = pd.read_csv("../data/raw/cota_2024_laps.csv")

df.head()
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
compound_map = {
    "SOFT": 0,
    "MEDIUM": 1,
    "HARD": 2
}

df["CompoundEncoded"] = df["Compound"].map(compound_map)
```

```python
df["DriverEncoded"] = df["Driver"].astype("category").cat.codes
```

```python
race_laps = 56

df["FuelLoadProxy"] = race_laps - df["LapNumber"]
```

```python
df["TyreDegEffect"] = df["TyreLife"]
```

```python
df["SectorSum"] = df["Sector1Sec"] + df["Sector2Sec"] + df["Sector3Sec"]
```

```python
model_df = df[[
    "Driver",
    "DriverEncoded",
    "Compound",
    "CompoundEncoded",
    "TyreLife",
    "FuelLoadProxy",
    "Sector1Sec",
    "Sector2Sec",
    "Sector3Sec",
    "LapTimeSec"
]]
```

```python
os.makedirs('../data/processed', exist_ok=True)
model_df.to_csv("../data/processed/cota_2024_features.csv", index=False)
```
