```python
fuel_effect = 0.060

deg_soft = 0.066
deg_medium = 0.029
deg_hard = 0.016
```

```python
base_lap = 100
```

```python
def simulate_stint(compound, laps, start_lap):

    if compound == "SOFT":
        deg = deg_soft
    elif compound == "MEDIUM":
        deg = deg_medium
    else:
        deg = deg_hard

    total_time = 0

    for i in range(laps):

        tyre_age = i
        fuel = start_lap + i

        lap_time = (
            base_lap
            + deg * tyre_age
            - fuel_effect * fuel
        )

        total_time += lap_time

    return total_time
```

```python
simulate_values = {}
simulate_map = {
"SOFT": [20, 10],
"MEDIUM": [20, 10],
"HARD": [20, 10]
}

for i in simulate_map.keys():
    simulate_values[i] = simulate_stint(i, simulate_map[i][0], simulate_map[i][1])
simulate_values
```

```python
import matplotlib.pyplot as plt

laps = range(20)

soft = []
medium = []
hard = []

for i in laps:

    soft.append(base_lap + deg_soft*i)
    medium.append(base_lap + deg_medium*i)
    hard.append(base_lap + deg_hard*i)

plt.plot(laps, soft, label="Soft")
plt.plot(laps, medium, label="Medium")
plt.plot(laps, hard, label="Hard")

plt.xlabel("Tyre Age")
plt.ylabel("Lap Time")
plt.title("Stint Pace Comparison")

plt.legend()
plt.show()
```
