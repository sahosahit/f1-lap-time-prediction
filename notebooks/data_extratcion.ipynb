```python
import os
import fastf1

CACHE_DIR = "../cache"
os.makedirs(CACHE_DIR, exist_ok=True)
fastf1.Cache.enable_cache(CACHE_DIR)
```

```python
session = fastf1.get_session(2024, 'United States', 'R')
session.load()
```

```python
laps = session.laps
laps.head()
```

```python
laps['LapTimeSec'] = laps['LapTime'].dt.total_seconds()

laps['Sector1Sec'] = laps['Sector1Time'].dt.total_seconds()
laps['Sector2Sec'] = laps['Sector2Time'].dt.total_seconds()
laps['Sector3Sec'] = laps['Sector3Time'].dt.total_seconds()
```

```python
laps_clean = laps[
    (laps['PitOutTime'].isna()) &
    (laps['PitInTime'].isna()) &
    (laps['TrackStatus'] == '1')
]
```

```python
dataset = laps_clean[[
    'Driver',
    'LapNumber',
    'Compound',
    'TyreLife',
    'Sector1Sec',
    'Sector2Sec',
    'Sector3Sec',
    'LapTimeSec'
]]
```

```python
os.makedirs('../data/raw', exist_ok=True)
dataset.to_csv('../data/raw/cota_2024_laps.csv', index=False)
```
